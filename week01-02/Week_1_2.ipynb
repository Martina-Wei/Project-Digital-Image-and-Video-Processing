Low Pass filter

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load IMG
img = cv2.imread('./bird.jpg')
img_cvt=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_cvt)
plt.show()
```

```python
ks = [30, 70]

for k in ks:
  fx = len(img_cvt)//k
  fy = len(img_cvt[0])//k
  kernel = np.ones((fx, fy), np.float32)/ (fx * fy)
  filteredImage = cv2.filter2D(img_cvt, -1, kernel)
  plt.imshow(filteredImage)
  plt.title('k = {}'.format(k))
  plt.show()
  
plt.imshow(img_cvt)
plt.title('original img')
plt.show()
```

High Pass filter

```python
hpf = np.array([[-1, -1, -1], [-1, 10, -1], [-1, -1, -1]], np.float32)
```

```python
filteredImage = cv2.filter2D(img_cvt, -1, hpf)
plt.imshow(filteredImage)
plt.show()
plt.imshow(img_cvt)
plt.show()
```
